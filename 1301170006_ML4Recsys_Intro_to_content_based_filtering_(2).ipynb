# Read data

```
import pandas as pd
import numpy as np

data = pd.read_csv('https://query.data.world/s/uikepcpffyo2nhig52xxeevdialfl7')
data = data[['Title','Genre','Director','Actors','Plot']]
pd.set_option('display.max_rows', None)
data
```

# Preprocessing

## Karena pada saat run akan sangat membutuhkan wktu yg lama

```
data = data.loc[0:19, "Title":"Actors"]
data
```

## menggabungkan nama actor, jadi dihapus spasi pada nama actor

```
actor = data["Actors"]
for i in range(len(actor)):
    actor[i] = actor[i].replace(" ","").replace(",", " ")

data["Actors"] = actor
```

#menghilangkan stopwords

```
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stopWords = stopwords.words('english')
print(stopWords)

row = []
for i in range(len(data)-1):
    text = " ".join(data.loc[i, "Title":"Actors"])
    row.append(text)

row
data1 = " ".join(row).split()
data1 = [x.replace(',', "") for x in data1]
data1 = [x.replace('.', "") for x in data1]
data1 = list(dict.fromkeys(data1))
data1 = [x.lower() for x in data1]

for word in data1:
    if word in stopWords:
        data1.remove(word)
print(len(data1))
data1
```

```
common_terms = ["of", "with", "without", "and", "or", "the", "a"]
def cleaningDocs(docs):
    listOfWord = docs
    listOfWord = [x.replace(',', "") for x in listOfWord]
    listOfWord = [x.replace('.', "") for x in listOfWord]
    listOfWord = [x.lower() for x in listOfWord]
    return listOfWord
```

# Vector Representation

```
def tf(word, docs):
    docs = cleaningDocs(" ".join(docs).split())
    listOfWord = list(" ".join(docs).split())
    # print(listOfWord)
    return listOfWord.count(word)/len(listOfWord)

def idf(df, word):
    documentsContainWord = 0
    for i in data.index:
        list = cleaningDocs(" ".join(data.loc[i, "Title":"Actors"]).split())
        if (word in list):
            documentsContainWord += 1
        # print("documentsContainWord", documentsContainWord)
    temp = len(data1)/documentsContainWord
    return np.log(temp)

def tfidf(tf,idf):
    return tf*idf

def build_vec_representation(x):
   
    all = []
    for i in range(len(data)):
        # print(i)
        baris = []
        docs = data.loc[i, "Title":"Actors"]
        for word in x:
            # print("word", word)
            term_freq = tf(word, docs)
            inv_doc_freq = idf(data, word)
            tf_idf = tfidf(term_freq, inv_doc_freq)
            baris.append(tf_idf)
        all.append(baris)
    
    result = pd.DataFrame(all, columns=x, index=data['Title'])

    return result

vector_representation = build_vec_representation(data1)
vector_representation
```

# Calculate the Similarity

```
from sklearn.metrics.pairwise import cosine_similarity
def cos_similarity(x):
  return cosine_similarity(x,x)

cos_simil = cosine_similarity(vector_representation)
cos_simil
```

```
cosine_matrix = pd.DataFrame(cos_simil, columns=data['Title'], index=data['Title'])
cosine_matrix
```

# Recommended

```
def recommend_film(film, cosine_matrix, k):
  
    list = [i for i in cosine_matrix.index]
    if film in list:
        list.remove(film)
    get_row = cosine_matrix.loc[film, list]
    get_row.sort_values(ascending=False, inplace=True)
    return get_row.head(k)

print(recommend_film("Se7en", cosine_matrix, 3))
```
